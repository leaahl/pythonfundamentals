```python
import pandas as pd
employees = pd.DataFrame({'name' : ['Susan', 'Kia', 'Sharvari', 'Nutan', 'Akash', 'Shadab', 'Neha', 'Karan'],'work_exp' : [10, 12, 11, 7.5, 6.5, 4.5, 9,15],'domain' : ['banking', 'hospitality', 'hospitality', 'insurance', 'banking', 'banking', 'network', 'insurance'],'skills' : ['C | C# | Java | Python','C# | Java | Python | R','Python | R |SAS |Tableau',
'C# | Java | Python | R |SAS', 'C# | Java | Python | R','C# | Java| Python', ' R|SAS|Python', ' R|SAS|Python'],
'SalBracket' : ['4500 - 5000', '4500 - 5000','4500 - 5000',
'2000 - 2500', '3500 - 4000', '2000 - 2500', '6000', '6000-6500'],'emp_id' : ['EMP003','EMP006','EMP004','EMP007','EMP001','EMP002','EMP008','EMP005']})

other_info = pd.DataFrame({'emp_id' : ['EMP001', 'EMP003', 'EMP005', 'EMP009', 'EMP011'],
'name' : ['Akash','Susi', 'Karan', 'Dean', 'Kia'], 'age':[26, 42, 55, 34, 28],
'hometown' : ['Delhi', 'Lucknow', 'Trichy', 'Hyderabad', 'Trivandrum']})

dep = pd.DataFrame({'EID' : ['EMP003', 'EMP009', 'EMP010'],
'Dept' : ['Testing', 'Analysis', 'HR']})
```

```python
employees
```

```python
other_info
```

```python
dep
```

```python
pd.merge(employees, other_info) #by default this is an inner join
```

```python
pd.merge(employees, other_info, how="inner"  ) #better to specify the inner so that it's clearer
```

```python
pd.merge(employees, other_info, how="inner", on="emp_id") #now i have 3 records because i only join on emp_id. by default it will join on both the common columns, name & empt_id
```

#### The suffix is also by default _x and _y, we can change that : 

```python
pd.merge(employees, other_info, how="inner", on="emp_id", suffixes = ('_emp', '_other'))
```

```python
pd.merge(other_info, dep, left_on="emp_id", right_on="EID" ) #if there are no common column headers
```

```python
other_info
```

```python
dep
```

```python
pd.merge(other_info,dep,how='left',left_on='emp_id',right_on='EID') #merging only on left join - all tables on other_info and dep that matches other_info records
```

```python
pd.merge(other_info,dep,how='right',left_on='emp_id',right_on='EID') #right join to the dep table
```

```python
pd.concat([other_info, dep]) #just concats and stacks 2 data frames on top of each other
```

```python
pd.concat([other_info, dep, employees])
```
